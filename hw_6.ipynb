```python
import numpy as np
from scipy.optimize import linprog

A = np.array([
    [6, 0, 6, 0, 1, 8, 6, 3, 5, 3],
    [6, 6, 3, 8, 0, 1, 4, 1, 0, 5],
    [7, 4, 9, 2, 0, 3, 3, 0, 7, 9]
], dtype=float)

b = np.array([18, 7, 19], dtype=float)
c = np.ones(10)

result = linprog(c, A_eq=A, b_eq=b, bounds=[(0, None)] * 10, method="highs")

print("Success:", result.success)
print("Minimizer x:", result.x)
print("Minimum value:", result.fun)
```
